```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 3. Histograms
plt.figure(figsize=(12,5))
```

```python
plt.subplot(1,2,1)
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
```

```python
plt.subplot(1,2,2)
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Fare Distribution")

```

```python
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare Distribution by Class")
```

```python
plt.subplot(1,2,2)
sns.boxplot(x="Sex", y="Age", data=df)
plt.title("Age Distribution by Gender")
```

```python
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.subplot(1,3,1)
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
```

```python
plt.subplot(1,3,2)
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
```

```python
plt.subplot(1,3,3)
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Class")
```

```python
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 7. Pairplot (Age, Fare, Pclass vs Survived)
sns.pairplot(df[["Survived", "Age", "Fare", "Pclass"]].dropna(), hue="Survived", diag_kind="hist")
plt.show()
```

```python
# 8. Summary observations (you can expand these manually after viewing plots)
print(" Quick Observations:")
print(" Females show higher survival rates compared to males.")
print(" Passengers in higher classes (1st > 2nd > 3rd) had better survival chances.")
print(" Younger passengers (children) survived more often.")
print(" Fare is highly skewed with many outliers at the higher end.")
print(" Strong correlation between Pclass and Fare (higher class paid more).")
```
